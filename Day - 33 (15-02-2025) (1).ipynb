## Pre Processing

```python
df=pd.get_dummies(titanic,dtype='int')
df.head()
```

## Apriori Algorithm

```python
titanic.shape[0]
```

```python
import warnings
warnings.filterwarnings('ignore')
frequent_itemsets=apriori(df,min_support=0.1,use_colnames=True) #support value change accornding to domain value requirement
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets, metric="lift", min_threshold=1, num_itemsets=len(frequent_itemsets))
rules
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```
