```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
cars[['HP','WT','VOL','SP']]=sc.fit_transform(cars[['HP','WT','VOL','SP']])
```

```python
np.round(cars.describe(),3)
```

```python
import statsmodels.formula.api as smf
ml_vol=smf.ols('MPG~VOL',data = cars).fit()
np.round(ml_vol.rsquared,2) # RSQUARE IS USED TO KNOW WHETHER YOUR MODEL IS GOOD OR BAD, IF IT IS NEAR TO 1 IT IS A GOOD MODEL
```

```python
ml_vol=smf.ols('MPG~HP',data = cars).fit()
np.round(ml_vol.rsquared,2)
```

```python
ml_vol=smf.ols('MPG~WT',data = cars).fit()
np.round(ml_vol.rsquared,2)
```

```python
ml_vol=smf.ols('MPG~SP',data = cars).fit()
np.round(ml_vol.rsquared,2)
```

```python
model = smf.ols('MPG~WT+SP+HP+VOL',data=cars).fit()
```

```python
np.round(model.rsquared,2)
```
